```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(df[['CPU_frequency']],df['Price'])
lm
```

```python
Yhat=lm.predict(df[['CPU_frequency']])
Yhat[0:5]
```

```python
width=12
height=10
plt.figure(figsize=(width,height))
ax1=sns.distplot(df['Price'],hist=False,label="Actual Price",color="r")
sns.distplot(Yhat,color="b",hist=False,label="fitted value",ax=ax1)
plt.title('Fitted value vs Actual Value')
plt.xlabel('Price(in dollars)')
plt.ylabel('proportion of laptops')
plt.legend(['Actual Value','Predicted Value'])
plt.show()
```

```python
print("The mean squared value is:",mean_squared_error(df['Price'],Yhat))
```

```python
lm.score(df[['CPU_frequency']],df['Price'])
```

```python
z=df[['CPU_frequency','RAM_GB','Storage_GB_SSD','CPU_core','OS','GPU','Category']]
y=df['Price']
lm.fit(z,y)
Y1hat=lm.predict(z)
Y1hat[0:5]
```

```python
plt.figure(figsize=(width,height))
ax1=sns.distplot(df['Price'],hist=False,label="Actual Value",color="r")
sns.distplot(Y1hat,hist=False,label="Fitted value",color="b")
plt.legend(['Actual Value','Predicted Value'])
plt.xlabel('Actual Value')
plt.ylabel('Fitted Value')
plt.title('Actual Value vs Predicted Value')
```

```python
mean_squared_error(y,Y1hat)
```

```python
lm.score(z,y)
```

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
x=df['CPU_frequency']
y=df['Price']
f=np.polyfit(x,y,3)
p=np.poly1d(f)
p
```

```python
PlotPolly(p,x,y,df['CPU_frequency'])
```

```python
mean_squared_error(df['Price'],p(x))
```

```python
r2_score(y,p(x))
```
