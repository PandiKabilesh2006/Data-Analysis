```python
''' 1.What is RobustScaler?
A Robust Scaler is a data preprocessing technique used in 
machine learning to scale features while minimizing the 
influence of outliers. Unlike standard scaling methods 
(like StandardScaler or MinMaxScaler), which can be 
heavily affected by extreme values, the Robust Scaler 
uses the median and the interquartile range (IQR) to 
transform the data.'''
```

```python
from sklearn.preprocessing import RobustScaler
robust=RobustScaler()
data=[[ 1., -2.,  2.],[ -2.,  1.,  3.],[ 4.,  1., -2.]]
robust.fit_transform(data)
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
data=[["Why"],["so"],["serious"],["?"]]
encoder.fit_transform(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data=[[5],[10],[12],[14],[18],[23],[49]]
scaler.fit_transform(data)
```

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
data=[[5],[10],[12],[14],[18],[23],[49]]
scaler.fit_transform(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data=[[ 1., -1.,  2.],[ 2.,  0.,  0.],[ 0.,  1., -1.]]
scaler.fit_transform(data)
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler=MaxAbsScaler()
X = [[ 1., -1.,  2.],[ 2.,  0.,  0.],[ 0.,  1., -1.]]
scaler.fit_transform(X)
```
